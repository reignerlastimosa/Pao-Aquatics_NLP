```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd
import time
```

```python
url = 'https://shopee.ph/buyer/42538878/rating'
chrome_driver = '/Users\R/Desktop/Web Scraper/chromedriver'
driver = webdriver.Chrome(chrome_driver)

driver.get('https://shopee.ph/buyer/42538878/rating')
```

```python
x=[]
def automate():
    time.sleep(5)    
    for customer_feedback in driver.find_elements_by_class_name('shopee-product-rating__main'):
        username = customer_feedback.find_element_by_class_name('shopee-product-rating__author-name').text
        date=customer_feedback.find_element_by_class_name('shopee-product-rating__time').text
        product=customer_feedback.find_element_by_class_name('KtBvER').text
        feedback=customer_feedback.find_element_by_class_name('shopee-product-rating__content').text   
        content={'username':username, 'date':date, 'product':product,'feedback':feedback}
        if date=='':
            return
        else:
            x.append(content)

```

```python
for i in range(10): #repeat next page
    time.sleep(2)
    automate()
    WebDriverWait(driver, 20).until(EC.element_to_be_clickable((By.CSS_SELECTOR, 'button.shopee-icon-button.shopee-icon-button--right'))).click()


```

```python
data = pd.DataFrame(data=x)
data
```
